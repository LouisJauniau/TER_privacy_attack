
# TER — Notebook YData Profiling

Ce notebook sert à **présenter les jeux de données** et **quelques résultats** du projet TER autour de :
- l'anonymisation,
- la linkage attack,
- la membership inference attack (MIA).

L'idée est de se servir de **YData Profiling** pour explorer visuellement les datasets, puis d'ajouter quelques tableaux maison pour résumer les métriques importantes.

## Ce que montre ce notebook
1. aperçu des fichiers utilisés ;
2. comparaison simple entre dataset publié et dataset anonymisé ;
3. profils YData des datasets principaux ;
4. résumé des métriques d'anonymisation, de linkage et de MIA ;
5. export optionnel des rapports HTML.

## Remarque pratique
Le notebook est pensé pour être placé **à la racine du projet** `ter_benchmark_starter_branch_27_03`.
S'il est lancé ailleurs, il essaie quand même de retrouver automatiquement le dossier du projet.


```python

# Décommente cette cellule si besoin d'installer les dépendances dans ton environnement Jupyter.
# %pip install -U pandas matplotlib ydata-profiling[notebook] ipywidgets
```

```python

from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

try:
    from ydata_profiling import ProfileReport
except ImportError as e:
    raise ImportError(
        "Le package 'ydata-profiling' n'est pas installé. "
        "Décommente la cellule d'installation au-dessus puis relance le kernel."
    ) from e

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 200)
```

```python

PROJECT_NAME = "ter_benchmark_starter_branch_27_03"

def find_project_root() -> Path:
    candidates = [
        Path.cwd(),
        Path.cwd() / PROJECT_NAME,
        Path.cwd().parent,
        Path.cwd().parent / PROJECT_NAME,
    ]
    for candidate in candidates:
        if (candidate / "data").exists() and (candidate / "outputs").exists():
            return candidate.resolve()
    raise FileNotFoundError(
        "Impossible de localiser la racine du projet. "
        "Place ce notebook à la racine du dépôt ou adapte PROJECT_ROOT manuellement."
    )

PROJECT_ROOT = find_project_root()
REPORTS_DIR = PROJECT_ROOT / "outputs" / "reports_ydata"
REPORTS_DIR.mkdir(parents=True, exist_ok=True)

print("PROJECT_ROOT =", PROJECT_ROOT)
print("REPORTS_DIR  =", REPORTS_DIR)
```

```python

# =========================
# FICHIERS UTILISÉS
# =========================

PATHS = {
    "original_with_id": PROJECT_ROOT / "data" / "adult_with_record_id.csv",
    "published_before_anonymization": PROJECT_ROOT / "outputs" / "prepared_data" / "adult_mia_age-sex-race.published.csv",
    "anonymized_public": PROJECT_ROOT / "outputs" / "anonymized" / "adult_mia_age-sex-race__k_5__l_2__t_None__supp_10__arx.csv",
    "attacker_base": PROJECT_ROOT / "outputs" / "prepared_data" / "adult_mia_age-sex-race.attacker_base.csv",
    "mia_targets": PROJECT_ROOT / "outputs" / "mia_targets" / "adult_mia_age-sex-race.targets.csv",
    "anonymization_metrics": PROJECT_ROOT / "outputs" / "metrics" / "adult_mia_age-sex-race__k_5__l_2__t_None__supp_10__arx.json",
    "mia_summary": PROJECT_ROOT / "outputs" / "attacks" / "mia" / "adult_mia_age-sex-race__k_5__l_2__t_None__supp_10__arx__mia_age-sex-race-native-country__targets_1000__seed_42" / "summary.json",
    "linkage_summary": PROJECT_ROOT / "outputs" / "attacks" / "linkage" / "qi_age-sex-race-marital-status-native-country__k_5__l_2__t_None__supp_10__arx__known_sex-marital-status-occupation-education-relationship-race-age-native-country__targets_3__seed_42" / "summary.json",
    "linkage_privjedai_summary": PROJECT_ROOT / "outputs" / "attacks" / "linkage_privjedai" / "qi_age-sex-race-marital-status-native-country__k_5__l_2__t_None__supp_10__arx__privjedai__known_sex-marital-status-occupation-education-relationship-race-age-native-country__targets_3__seed_42" / "summary.json",
}

paths_df = pd.DataFrame(
    [{"name": name, "path": str(path), "exists": path.exists()} for name, path in PATHS.items()]
)
display(paths_df)

missing = paths_df.loc[~paths_df["exists"], "name"].tolist()
if missing:
    print("Fichiers manquants :", missing)
else:
    print("Tous les fichiers attendus sont présents.")
```

```python

# =========================
# CHARGEMENT DES DONNÉES
# =========================

df_original = pd.read_csv(PATHS["original_with_id"])
df_published = pd.read_csv(PATHS["published_before_anonymization"])
df_anonymized = pd.read_csv(PATHS["anonymized_public"])
df_attacker = pd.read_csv(PATHS["attacker_base"])
df_targets = pd.read_csv(PATHS["mia_targets"])

print("original_with_id               ->", df_original.shape)
print("published_before_anonymization ->", df_published.shape)
print("anonymized_public              ->", df_anonymized.shape)
print("attacker_base                  ->", df_attacker.shape)
print("mia_targets                    ->", df_targets.shape)
```

```python

# Aperçu rapide

display(Markdown("## Aperçu du dataset publié avant anonymisation"))
display(df_published.head())

display(Markdown("## Aperçu du dataset anonymisé public"))
display(df_anonymized.head())

display(Markdown("## Aperçu de la base de l'attaquant"))
display(df_attacker.head())

display(Markdown("## Aperçu des targets MIA"))
display(df_targets.head())
```

```python

# =========================
# TABLEAU DE COMPARAISON SIMPLE
# =========================

def dataset_overview(df: pd.DataFrame, name: str) -> dict:
    return {
        "dataset": name,
        "n_rows": len(df),
        "n_cols": df.shape[1],
        "missing_values_total": int(df.isna().sum().sum()),
        "duplicated_rows": int(df.duplicated().sum()),
    }

comparison_df = pd.DataFrame([
    dataset_overview(df_original, "original_with_id"),
    dataset_overview(df_published, "published_before_anonymization"),
    dataset_overview(df_anonymized, "anonymized_public"),
    dataset_overview(df_attacker, "attacker_base"),
    dataset_overview(df_targets, "mia_targets"),
])

display(comparison_df)
```

```python

# Comparaison de cardinalité sur quelques attributs intéressants

interesting_cols = [
    col for col in ["age", "sex", "race", "native-country", "income"]
    if col in df_published.columns and col in df_anonymized.columns
]

cardinality_rows = []
for col in interesting_cols:
    cardinality_rows.append({
        "column": col,
        "published_nunique": df_published[col].nunique(dropna=False),
        "anonymized_nunique": df_anonymized[col].nunique(dropna=False),
    })

cardinality_df = pd.DataFrame(cardinality_rows)
display(cardinality_df)
```


## YData Profiling

Les cellules suivantes génèrent des profils YData.

Pour éviter que le notebook soit trop lourd, le paramètre `minimal=True` est activé par défaut.
Tu peux le passer à `False` si tu veux un rapport plus riche.


```python

def build_profile(df: pd.DataFrame, title: str, minimal: bool = True):
    return ProfileReport(
        df,
        title=title,
        explorative=True,
        minimal=minimal,
        progress_bar=True,
    )
```

```python

# Profil YData : dataset publié avant anonymisation
profile_published = build_profile(
    df_published,
    title="TER - Published dataset before anonymization",
    minimal=True,
)
profile_published.to_notebook_iframe()
```

```python

# Profil YData : dataset anonymisé public
profile_anonymized = build_profile(
    df_anonymized,
    title="TER - Public anonymized dataset",
    minimal=True,
)
profile_anonymized.to_notebook_iframe()
```

```python

# Profil YData : base de l'attaquant
profile_attacker = build_profile(
    df_attacker,
    title="TER - Attacker knowledge base",
    minimal=True,
)
profile_attacker.to_notebook_iframe()
```

```python

# Profil YData : cibles MIA
profile_targets = build_profile(
    df_targets,
    title="TER - MIA targets",
    minimal=True,
)
profile_targets.to_notebook_iframe()
```

```python

# Export HTML des rapports

profile_published.to_file(REPORTS_DIR / "published_before_anonymization_profile.html")
profile_anonymized.to_file(REPORTS_DIR / "anonymized_public_profile.html")
profile_attacker.to_file(REPORTS_DIR / "attacker_base_profile.html")
profile_targets.to_file(REPORTS_DIR / "mia_targets_profile.html")

print("Rapports HTML exportés dans :", REPORTS_DIR)
```

```python

# =========================
# MÉTRIQUES D'ANONYMISATION
# =========================

def load_json(path: Path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

anonym_metrics = load_json(PATHS["anonymization_metrics"])

anonym_table = pd.DataFrame([{
    "experiment_id": anonym_metrics.get("experiment_id"),
    "status": anonym_metrics.get("status"),
    "anonymization_time_ms": anonym_metrics.get("anonymization_time_ms"),
    "number_of_equivalence_classes": anonym_metrics.get("number_of_equivalence_classes"),
    "average_equivalence_class_size": anonym_metrics.get("average_equivalence_class_size"),
    "min_equivalence_class_size": anonym_metrics.get("min_equivalence_class_size"),
    "max_equivalence_class_size": anonym_metrics.get("max_equivalence_class_size"),
    "number_of_suppressed_records": anonym_metrics.get("number_of_suppressed_records"),
    "ambiguity_metric": anonym_metrics.get("ambiguity_metric"),
    "average_class_size_metric": anonym_metrics.get("average_class_size_metric"),
}])

display(anonym_table)

transformations = anonym_metrics.get("transformations", {})
if transformations:
    display(Markdown("### Niveaux de transformation ARX"))
    display(pd.DataFrame(list(transformations.items()), columns=["attribute", "level"]))
```

```python

# =========================
# RÉSUMÉ MIA
# =========================

mia_summary = load_json(PATHS["mia_summary"])

mia_table = pd.DataFrame([{
    "attack_id": mia_summary.get("attack_id"),
    "n_targets": mia_summary.get("n_targets"),
    "n_members": mia_summary.get("n_members"),
    "n_non_members": mia_summary.get("n_non_members"),
    "tp": mia_summary.get("tp"),
    "tn": mia_summary.get("tn"),
    "fp": mia_summary.get("fp"),
    "fn": mia_summary.get("fn"),
    "accuracy": mia_summary.get("accuracy"),
    "precision": mia_summary.get("precision"),
    "recall": mia_summary.get("recall"),
    "f1": mia_summary.get("f1"),
    "member_recall": mia_summary.get("member_recall"),
    "non_member_true_negative_rate": mia_summary.get("non_member_true_negative_rate"),
    "member_avg_compatible_candidate_count": mia_summary.get("member_avg_compatible_candidate_count"),
    "non_member_avg_compatible_candidate_count": mia_summary.get("non_member_avg_compatible_candidate_count"),
    "member_avg_best_score": mia_summary.get("member_avg_best_score"),
    "non_member_avg_best_score": mia_summary.get("non_member_avg_best_score"),
}])

display(mia_table.T.rename(columns={0: "value"}))
```

```python

# =========================
# RÉSUMÉ LINKAGE
# =========================

linkage_summary = load_json(PATHS["linkage_summary"])
linkage_priv_summary = load_json(PATHS["linkage_privjedai_summary"])

linkage_compare = pd.DataFrame([
    {
        "version": "linkage_classique",
        "n_targets": linkage_summary.get("n_targets"),
        "target_survival_rate": linkage_summary.get("target_survival_rate"),
        "true_record_in_compatible_rate": linkage_summary.get("true_record_in_compatible_rate"),
        "true_record_in_best_rate": linkage_summary.get("true_record_in_best_rate"),
        "unique_exact_reidentification_rate": linkage_summary.get("unique_exact_reidentification_rate"),
        "avg_compatible_candidate_count": linkage_summary.get("avg_compatible_candidate_count"),
        "median_compatible_candidate_count": linkage_summary.get("median_compatible_candidate_count"),
        "avg_best_candidate_count": linkage_summary.get("avg_best_candidate_count"),
        "compatible_sensitive_disclosure_rate": linkage_summary.get("compatible_sensitive_disclosure_rate"),
        "best_sensitive_disclosure_rate": linkage_summary.get("best_sensitive_disclosure_rate"),
    },
    {
        "version": "linkage_privjedai",
        "n_targets": linkage_priv_summary.get("n_targets"),
        "target_survival_rate": linkage_priv_summary.get("target_survival_rate"),
        "true_record_in_compatible_rate": linkage_priv_summary.get("true_record_found_rate"),
        "true_record_in_best_rate": linkage_priv_summary.get("top1_reidentification_rate"),
        "unique_exact_reidentification_rate": linkage_priv_summary.get("top1_reidentification_rate"),
        "avg_compatible_candidate_count": linkage_priv_summary.get("avg_match_count"),
        "median_compatible_candidate_count": linkage_priv_summary.get("median_match_count"),
        "avg_best_candidate_count": None,
        "compatible_sensitive_disclosure_rate": linkage_priv_summary.get("certainty_sensitive_inference_rate"),
        "best_sensitive_disclosure_rate": linkage_priv_summary.get("avg_true_sensitive_weighted_share"),
    },
])

display(linkage_compare)
```

```python

# =========================
# PETITS GRAPHIQUES MAISON
# =========================

# 1) Cardinalité publiée vs anonymisée pour quelques colonnes
if not cardinality_df.empty:
    ax = cardinality_df.set_index("column")[["published_nunique", "anonymized_nunique"]].plot(kind="bar", figsize=(10, 5))
    ax.set_title("Nombre de valeurs distinctes : publié vs anonymisé")
    ax.set_ylabel("Nombre de valeurs distinctes")
    ax.set_xlabel("Colonne")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# 2) Quelques métriques MIA
mia_plot_df = pd.DataFrame({
    "metric": ["accuracy", "precision", "recall", "f1", "member_recall"],
    "value": [
        mia_summary.get("accuracy"),
        mia_summary.get("precision"),
        mia_summary.get("recall"),
        mia_summary.get("f1"),
        mia_summary.get("member_recall"),
    ]
})

ax = mia_plot_df.plot(x="metric", y="value", kind="bar", figsize=(8, 4), legend=False)
ax.set_title("MIA - métriques principales")
ax.set_ylabel("Valeur")
ax.set_xlabel("Métrique")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

# 3) Taille moyenne des ensembles de candidats en linkage
linkage_plot = linkage_compare[["version", "avg_compatible_candidate_count"]].dropna()
ax = linkage_plot.plot(x="version", y="avg_compatible_candidate_count", kind="bar", figsize=(7, 4), legend=False)
ax.set_title("Linkage - taille moyenne des candidats")
ax.set_ylabel("Nombre moyen de candidats")
ax.set_xlabel("Version")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```


## Comment présenter ce notebook à l'oral ou dans le rapport

Tu peux structurer l'explication comme ça :

1. **Avant anonymisation** : montrer la structure du dataset publié et ses distributions principales.
2. **Après anonymisation** : montrer les mêmes variables et faire observer la baisse de granularité.
3. **Point de vue attaquant** : montrer ce qu'il voit réellement dans `attacker_base` et dans les `targets`.
4. **Résultats d'anonymisation** : rappeler le nombre de classes d'équivalence, leur taille moyenne, et la suppression.
5. **Résultats MIA** : insister sur accuracy, recall côté membres, et taille moyenne des ensembles compatibles.
6. **Résultats linkage** : comparer la version classique et la version privJedAI sur la taille moyenne des candidats et les taux de ré-identification.
